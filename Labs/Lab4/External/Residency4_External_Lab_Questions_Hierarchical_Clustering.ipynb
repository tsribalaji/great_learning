# Hierarchical Clustering

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(101)  # for repeatability of this dataset
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

### 1. Combine all three arrays a,b,c into a dataframe
#### Hint: Use concatenate function from numpy and axis=0

```python
dataset = np.concatenate([a, b, c], axis=0)
dataset
```

### 2. Use scatter plot to print all the 3 distributions

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#sns.scatterplot(data=dataset, palette='spring')
plt.scatter(dataset[:, 0], dataset[:, 1], s=30, cmap='prism')
```

### 3. Find linkage matrix
#### Use ward as linkage metric and distance as Eucledian. 
#### From scipy.cluster.hierarchy import dendrogram, linkage
#### Generate the linkage matrix

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linkage_mt= linkage(dataset, method='ward', metric='euclidean')
```

### 4. Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(10, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(linkage_mt, leaf_rotation=90.,color_threshold = 40, leaf_font_size=8. )
plt.tight_layout()
```

### 5. Recreate the dendrogram for last 12 merged clusters
#### Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

```python
plt.figure(figsize=(10, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(linkage_mt, p=12, truncate_mode='lastp', leaf_rotation=90.,color_threshold = 40, leaf_font_size=8. )
plt.tight_layout()
```

### 6.From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

From the above dendogram, there are 3 groups under blue line. So we will choose the 3 clusters for clustering the data.

### 7. Use distance measure and fcluster function to cluster the data into 3 different groups
#### Hint: From scipy.cluster.hierarchy Import fcluster and use criterion as "distance". Print clusters to view the clustered data

```python
from scipy.cluster.hierarchy import fcluster
cluster = fcluster(linkage_mt, t=50, depth=2, criterion='distance')
```

### 8. Use matplotlib to visually observe the clusters

```python
plt.scatter(dataset[:, 0], dataset[:, 1], c=cluster, s=30, cmap='prism')
```
